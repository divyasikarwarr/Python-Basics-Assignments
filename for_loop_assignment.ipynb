## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
n = int(input("Enter a number:"))

if n < 2:
    is_prime = False
elif n == 2:
    is_prime = True
else:
    is_prime = True
    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

if is_prime:
    print(f"{n} is a prime number.")
else:
    print(f"{n} is not a prime number.")
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
n = int(input("Enter the upper limit N:"))

print(f"Prime numbers between 1 and {n} are:")

for num in range(2, n + 1):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)
```

## Question 18 (Hard)

**Title:** First Number Divisible by 7 and 11

### Problem Statement:
Take N as input and find the first number between 1 and N that is divisible by both 7 and 11. Stop after finding it.

### Approach:
Use logical AND and stop the loop using `break`.


```python
n = int(input("Enter the upper limit N:"))

found = False
for i in range(1, n + 1):
    if i % 7 == 0 and i % 11 == 0:
        print(f"The first number divisible by both 7 and 11 between 1 and {n} is: {i}")
        found = True
        break

if not found:
    print(f"No number found between 1 and {n} divisible by both 7 and 11.")
```

## Question 19 (Hard)

**Title:** Number Pattern

### Problem Statement:
Take N as input and print a right-angled number pattern using nested loops.

Example:
1
12
123

### Approach:
Use an outer loop for rows and an inner loop for printing numbers.


```python
n = int(input("Enter the number N:"))

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```
