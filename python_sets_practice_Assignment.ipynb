### Q3
Create an empty set and add numbers from 1 to 5 using a loop.

````python
# write your code here

s = set()
for i in range(1,6):
    s.add(i)
print(s)
````

### Q4
Why does this code give an error? Fix it.

```python

s.add(10)
```

````python
# write your code here

s = set()
s.add(10)
````

The code throws an error because we have used curly brackets to form an empty set. However, if we use {} to form an empty set, it becomes an empty dictionary not a Set. we have corrected the code by using s=set().

### Q5
Given:
```python
A = {2, 4, 6, 8}
B = {4, 8, 12}
```
Find Union, Intersection, A-B

````python
# write your code here

A = {2, 4, 6, 8}
B = {4, 8, 12}
print(A.union(B))
print(A.intersection(B))
print(A.difference(B))
````
